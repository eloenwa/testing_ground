```python

Stock = 'CAR'


url = 'https://markets.businessinsider.com/stocks/{}-stock'
get_financials_url = url.format(Stock,Stock)
r = requests.get(get_financials_url)
data =r.json
```

```python
soup = BeautifulSoup(r.content, "html.parser") 
content = soup.find_all(class_='push-data aktien-big-font text-nowrap no-padding-at-all')
content
```

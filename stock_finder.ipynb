```python
import import_ipynb
import os
import numpy as np 
import pandas as pd 
import yfinance as yf
import time
import import_ipynb
import json

def top_25():
    from sma_dates import start_stop_dates
    start_date, stop_date = start_stop_dates()
    print('start date', start_date)
    print('Stop date', stop_date)

    from clear_screen import Clear_screen

    stock_universe = pd.read_csv('nasdaq-listed-symbols_csv.csv')

    def get_stock_rating(stock):
        try:
            stock_ =yf.Ticker(stock)
            x=stock_.info['trailingEps']  
        except:
            x= 0  
        try:
             y = stock_.info['forwardPE']
        except:
             y=0
        try:
            z =stock_.info['marketCap']
        except:
            z=0
        try:
            a=stock_.info['dividendYield']
        except:
            a=0
        return([stock,x,y,z,a])


    simple_universe = []
    size = len(stock_universe['Symbol'])
    k = 0.0
    for symbol in stock_universe['Symbol']:
        data = get_stock_rating(str(symbol))
        simple_universe.append(data)
        k=k+1
        print(k,'of',size)
        Clear_screen()

    my_txt = open('simple_universe.txt','w')
    for things in simple_universe:
        my_txt.write(str(things)+'\n')
    my_txt.close()

    k=0.0
    lean_universe =[]
    for things in simple_universe:
        if things[1]==0 and things[2]==0:
            p=0
        else:
            lean_universe.append(things)


    k=0.0
    leaner_universe =[]
    for things in lean_universe:
        if things[3] == None:
            p= 0
        elif things[3] < 1*10**9:
            p=0
        elif things[3] > 1*10**9 and things[3] < 10*10**9:
            leaner_universe.append(things)
        else:
            p=0

    my_txt = open('leaner_universe.txt','w')
    for things in leaner_universe:
        my_txt.write(str(things)+'\n')
    my_txt.close()

    my_txt = open('leaner2_universe.txt','w')

    leaner2_universe =[]
    for things in leaner_universe:
        if things[1] == None:
            p= 0
        elif things[1] < 0:
            p=0
        else:
            leaner2_universe.append(things)
    for things in leaner2_universe:
        my_txt.write(str(things)+'\n')
    my_txt.close()

    my_txt = open('leaner2_universe.txt','w')

    leaner2_universe =[]
    for things in leaner_universe:
        if things[1] == None:
            p= 0
        elif things[1] < 0:
            p=0
        else:
            leaner2_universe.append(things)
    for things in leaner2_universe:
        my_txt.write(str(things)+'\n')
    my_txt.close()

    def takeSecond(elem):
        return elem[1]

    leaner2_universe.sort(key=takeSecond,reverse = True)


    def buy_sell(stock):
        try:
            data2= yf.download(stock,start=stop_date,end=start_date)
            sma = []
            tt =5
            sma=data2['Open'].rolling(tt).mean()    
            #fig, ax = plt.subplots( nrows=1, ncols=1 )  # create figure & 1 axis
            #ax.plot(sma)
            #ax.plot(data2['Open'])
            #fig.savefig(stock+'.png')   # save the figure to file
            #plt.close(fig)
        except:
            print('error:',stock)
        return()

    for watchlist in leaner2_universe:
        stock = watchlist[0]
        buy_sell(stock)

    def elo_score(stock):

        data = yf.Ticker(stock)
        stock = data.history()
        V_Avg = stock['Volume']

        sma = []
        tt =5
        sma=stock['Open'].rolling(tt).mean()

        elo_score =[]
        k=1.0
        for i in range(tt,len(sma)-1):


            result = (V_Avg[i]/10**6)*k*np.exp(k*(-sma[i]+stock['Open'][i])/sma[i])*(-sma[i]+stock['Open'][i])


            elo_score.append(result)

            k= k+1


        return((np.sum(elo_score)))

    elo_ranking = []
    time_1 =time.time()
    for data in leaner2_universe:
        result = elo_score(data[0])
        aa = [data[0],result]
        elo_ranking.append(aa)

    time_2 = time.time()
    print('elapsed time ',time_2-time_1)

    updated_elo_ranking = []
    for item in elo_ranking:
        if item[1]== None:
            p=0
        elif item[1] < 0:
            p=0
        else:
            updated_elo_ranking.append(item)

    updated_elo_ranking.sort(key=takeSecond,reverse = True)

    final_ranking = []
    k =0.0
    for items in updated_elo_ranking:

        final_ranking.append(items)
        if k == 10:
            break
        k=k+1

    
    with open('data2.JSON', 'w') as f:
        json.dump(final_ranking, f, ensure_ascii=False)
        
    with open('mini_universe.JSON', 'w') as f:
        json.dump(leaner2_universe, f, ensure_ascii=False)
        
    return
```

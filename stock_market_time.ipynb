```python
#check if the time is within the stock market open window 
def is_market_open():
    from datetime import datetime, time
    
    noow =(datetime.today())
    Market_open = time(9,10)
    Market_close = time(14,30)
    tme = noow.time()
    
    if tme > Market_open: 
        if tme < Market_close: 
            return True
        else:
            return False
    else: 
        return False
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import datetime
%matplotlib inline

```

```python
stock ='EHTH'
start = '2020-01-01'
end = '2020-03-18'
```

```python
def situation(stock,start,stop):
    data2= yf.download(stock,start=start,end=end)
    sma = []
    tt =5
    sma=data2['Open'].rolling(tt).mean()  
    pp = sma.index
    fig, ax = plt.subplots( nrows=1, ncols=1 )  # create figure & 1 axis
    ax.plot(sma)
    ax.plot(data2['Open']) 
    
    for i in range(tt,len(data2['Open'])):
        if data2['Open'][i] < sma[i]:
            print("sell", pp[i])
            action = "SELL"
        elif data2['Open'][i] > sma[i]:
            print("buy", pp[i])
            action="BUY"

    return action
```

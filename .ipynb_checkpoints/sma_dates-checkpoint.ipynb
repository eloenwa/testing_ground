```python
import datetime
from datetime import timedelta
dt = datetime.datetime.today()

def dt_2_date(dt):
    todays_date = str(str(dt.year)+'-'+str(dt.month)+'-'+str(dt.day))

    return todays_date

def start_stop_dates():
    dt = datetime.datetime.today()
    newdt = dt-timedelta(50)
    stop_date = dt_2_date(newdt)
    start_date = dt_2_date(dt)
    
    return start_date, stop_date
```

```python
import requests 
import json
from matplotlib import pyplot as plt
%matplotlib inline
from bs4 import BeautifulSoup
import re
import yfinance
import pandas as pd
import numpy as np 
import seaborn as sn 


endpoint_url= 'https://paper-api.alpaca.markets'
trading_id = 'PKRLHTOE9CXJ2ZB55OEC'
API_key = '2CyKQOltQNYejXNhTj7oKj3gmlaWQHJvDuMa/6eJ'
Account_url = "{}/v2/account".format(endpoint_url)

ORDERS_URL = '{}/v2/orders'.format(endpoint_url)

def get_account():
    r = requests.get(Account_url, headers={'APCA-API-KEY-ID':trading_id,'APCA-API-SECRET-KEY':API_key})
    user_data= r.json()
    return(user_data)

def create_order(symbol,qty,side,type1,time_in_force):
    endpoint_url= 'https://paper-api.alpaca.markets'
    ORDERS_URL = '{}/v2/orders'.format(endpoint_url)
    
    trading_id = 'PKRLHTOE9CXJ2ZB55OEC'
    API_key = '2CyKQOltQNYejXNhTj7oKj3gmlaWQHJvDuMa/6eJ'
    data = {
        "symbol":symbol,
        "qty": qty,
        "side": side,
        "type": type1,
        "time_in_force": time_in_force
    }

        
    r = requests.post(ORDERS_URL,json=data, headers={'APCA-API-KEY-ID':trading_id,'APCA-API-SECRET-KEY':API_key})
    return json.loads(r.content)

def get_portfolio(symbol,qty,side,type1,time_in_force):
    endpoint_url= 'https://paper-api.alpaca.markets'
    Portfolio_url = '{}/v1/positions'.format(endpoint_url)
    
    trading_id = 'PKRLHTOE9CXJ2ZB55OEC'
    API_key = '2CyKQOltQNYejXNhTj7oKj3gmlaWQHJvDuMa/6eJ'
    data = {
        "symbol":symbol,
        "qty": qty,
        "side": side,
        "type": type1,
        "time_in_force": time_in_force
    }

        
    r = requests.post(ORDERS_URL,json=data, headers={'APCA-API-KEY-ID':trading_id,'APCA-API-SECRET-KEY':API_key})
    return json.loads(r.content)

def Get_financials_2(Stock):
    url = 'https://markets.businessinsider.com/stocks/{}-stock'
    get_financials_url = url.format(Stock,Stock)
    r = requests.get(get_financials_url)
    #data =r.json
    
    soup = BeautifulSoup(r.content, "html.parser") 
    content = soup.find_all(class_='push-data aktien-big-font text-nowrap no-padding-at-all')
    ttty = str(content[0])
    #print(ttty)
    t23= ttty.split()
    
    t2=t23[len(t23)-1]
    temp = re.findall(r'\d+', t2) 
    #print(temp)
    if len(temp)==5:
        temp_1 = str(temp[2])+str(temp[3])
        #print(temp_1)
        temp[2]= temp_1
        temp[3]= temp[4]
        
    res = list(map(int, temp))
    realtime_stock_price= (res[2]+res[3]/100)
    #print(realtime_stock_price)
    #print(res)
    return(realtime_stock_price)
```

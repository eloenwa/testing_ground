```python
with open('data3.JSON') as f:
    data = json.load(f)
stock = data[2]
stock
```

```python

live_p = []

cases = ['sell','sell']
pp=0
k= 0
while True:
    while is_market_open() == True:
        ttt=Get_financials_2(stock)
        live_p.append(ttt)
        aaa= pd.DataFrame(live_p)
        try:
            sma=aaa.rolling(200).mean()
            sma_22 = aaa.rolling(500).mean()
            check_price=sma[0][len(sma)-1]
            Other_check = sma_22[0][len(sma_22)-1]


            if (check_price* 1.1) > Other_check:
                ttty=Get_financials_2(stock)
                if ttty *(1.1) > check_price:

                    #print('live price is:',ttt, 'sma:',check_price,'buy now', 'at ', (datetime.datetime.now().time()))
                    k = 1
                    cases[1] = "BUY"

            else:
                #print('live price is:',ttt, 'sma:',check_price,'sell now', 'at ', (datetime.datetime.now().time()))
                cases[1] = "SELL"

            if   ttt   < check_price:
                #print('live price is:',ttt, 'sma:',check_price,'sell now', 'at ', (datetime.datetime.now().time()))
                cases[1] = "SELL"   


            if k==1:
                #print('HWye')
                if cases[0] != cases[1]:
                    if cases[1] == "BUY":
                        create_order(stock,200,"buy","market","gtc")
                        print('BUY')
                        print('live price is:',ttt, 'sma:',check_price,'buy now', 'at ', (datetime.datetime.now().time()))
                    elif cases[1] == "SELL":
                        create_order(stock,200,"sell","market","gtc")
                        print('sell')
                        print('live price is:',ttt, 'sma:',check_price,'sell now', 'at ', (datetime.datetime.now().time()))
                        aaa =[]
                        live_p = []


        except:
            print('error')
        cases[0]=cases[1]
        sleep(1)
```

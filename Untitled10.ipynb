```python
from stock_loop_checker import Get_financials_2, create_order
from time import sleep
import pandas as pd
import datetime
```

```python


live_p = []

cases = ['sell','sell']
pp=0
k= 0
stock ="CAR"
while True:
    ttt=Get_financials_2(stock)
    live_p.append(ttt)
    aaa= pd.DataFrame(live_p)
    try:
        sma=aaa.rolling(20).mean()
        sma_22 = aaa.rolling(50).mean()
        check_price=sma[0][len(sma)-1]
        Other_check = sma_22[0][len(sma_22)-1]
        
        if   ttt > check_price:
            print('live price is:',ttt, 'sma:',check_price,'sell now', 'at ', (datetime.datetime.now().time()))
            cases[1] = "SELL"
        
        if check_price > Other_check:
            ttty=Get_financials_2(stock)
            if ttty > check_price:
        
                print('live price is:',ttt, 'sma:',check_price,'buy now', 'at ', (datetime.datetime.now().time()))
                k = 1
                cases[1] = "BUY"
            
        else:
            print('live price is:',ttt, 'sma:',check_price,'sell now', 'at ', (datetime.datetime.now().time()))
            cases[1] = "SELL"
            
            
        
            
        if k==1:
            print('HWye')
            if cases[0] != cases[1]:
                if cases[1] == "BUY":
                    create_order(stock,50,"sell","market","gtc")
                    print('BUY')
                elif cases[1] == "SELL":
                    create_order(stock,50,"buy","market","gtc")
                    print('sell')
            
        
    except:
        print('error')
    cases[0]=cases[1]
    sleep(2)
    
```
